```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.datasets import fetch_california_housing

from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
import pandas as pd
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
house_price_dataset = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X= np.hstack([house_price_dataset.values[::2, :], house_price_dataset.values[1::2, :2]])
y = house_price_dataset.values[1::2, 2]
```

```python
data=X
```

```python
columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
df=pd.DataFrame(data=data,columns=columns)
```

```python
df["price"]=y
```

```python
X=df.drop(["price"],axis=1)
y=df["price"]
```

splitting the data into Training data and Test data

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(y_train[211])
```

```python
model=XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
#accuracy for prediction on training data
training_data_predict=model.predict(X_train)
```

```python
print(training_data_predict)
```

```python
#R squared error
score_1=metrics.r2_score(y_train,training_data_predict)

#Mean Absolute Error
score_2=metrics.mean_absolute_error(y_train,training_data_predict)

print("R squared error:",score_1)
print("mean absolute error:",score_2)
```

```python
plt.scatter(y_train,training_data_predict)
plt.xlabel("Actual prices")
plt.ylabel("predicted prices")
plt.title("actual price vs prediced price")
plt.show()
```

```python
test_data_predict=model.predict(X_test)

score_1=metrics.r2_score(y_test,test_data_predict)
score_2=metrics.mean_absolute_error(y_test,test_data_predict)
```

```python
print("R squared error:",score_1)
print("mean absolute error",score_2)
```
